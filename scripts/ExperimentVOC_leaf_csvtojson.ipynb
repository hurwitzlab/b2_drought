```python
my_frame=graphlab.SFrame.read_csv(file_csv, delimiter="\t",column_type_hints=str)
my_frame
```

```python
for i in range(0,nb_samples):
    curr_slice=my_frame[i]
    
    line="{"
    #process the experiment description
    for i in range(len(Experiment_description)):
        Key=Experiment_description[i]
        if attribute_types[Key]=="int" or attribute_types[Key]=="float":
            attribute=str(curr_slice[Experiment_description[i]])
        elif attribute_types[Key]=="dateTime":
            if str(curr_slice[Experiment_description[i]])=="NA" or str(curr_slice[Experiment_description[i]])=="None":
                attribute="\"None\""
            else:
                attribute="{$date:\""+str(curr_slice[Experiment_description[i]])+"\"}"
        else:
            attribute="\""+str(curr_slice[Experiment_description[i]])+"\""
        line=line+"\""+Key+"\":"+attribute+","

    #print("#### "+line)
    
    #process the Results description
    line=line+"\"Results\":{"
    for i in range(len(Results)):
        Key=Results[i]
        if attribute_types[Key]=="int" or attribute_types[Key]=="float":
            attribute=str(curr_slice[Results[i]])
        elif attribute_types[Key]=="dateTime":
            if str(curr_slice[Results[i]])=="NA" or str(curr_slice[Results[i]])=="None":
                attribute="\"None\""
            else:
                attribute="{$date:\""+str(curr_slice[Results[i]])+"\"}"
        else:
            attribute="\""+str(curr_slice[Results[i]])+"\""
        line=line+"\""+Key+"\":"+attribute
        
        if i==len(Results)-1:
            line=line+"}"
        else:
            line=line+","
    #print("#### "+line)    
    
    line=line+"}"
    fout.write(line+"\n")
    print(line)
    
#fout.close()
```
