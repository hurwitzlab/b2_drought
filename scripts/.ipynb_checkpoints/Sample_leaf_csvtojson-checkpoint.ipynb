```python
import graphlab
```

```python
#File objects description
file_csv="samples_leaves.csv"
json_file="samples_leaves.json"
fout = open(json_file, "a")

nb_samples=51

Specimen_description=["sample_code", "sample_number_on_tag","sampler_size","replicate_ID","sun_exposure",
                      "leaf_condition_day_1-note_any_spots_or_damage","leaf_age",
                      "leaf_condition_day_of_collection-note_any_spots_or_damage","leaf_sample_ss_date","notes"]
Collection=["collection_date-time",
            "collection_protocol","collection_length_of_plastic_enclosure_in_days",
           "collection_treatment","collection_date-time_enclosure_placed"]
Location=["location_grid","location_host","location_tree_ID",
         "location_height_level","location_height_from_ground_in_m","location_closest_space_frame_level_from_0_to_18"]
```

```python
#file type description
attribute_types={
    "sample_code": "string",
    "sample_number_on_tag":"string",
    "sampler_size":"string",
    "replicate_ID":"string",
    "sun_exposure":"string",
    "leaf_condition_day_1-note_any_spots_or_damage":"string",
    "leaf_age":"string",
    "leaf_condition_day_of_collection-note_any_spots_or_damage":"string",
    "leaf_sample_ss_date":"dateTime",
    "notes":"string",
    "collection_date-time":"dateTime",
    "collection_protocol":"string",
    "collection_length_of_plastic_enclosure_in_days":"int",
    "collection_treatment":"string",
    "collection_date-time_enclosure_placed":"dateTime",
    "location_grid":"string",
    "location_host":"string",
    "location_tree_ID":"string",
    "location_height_level":"string",
    "location_height_from_ground_in_m":"float",
    "location_closest_space_frame_level_from_0_to_18":"int"
}
```

```python
my_frame=graphlab.SFrame.read_csv(file_csv, column_type_hints=str)
my_frame
```

```python


for i in range(0,nb_samples):
    curr_slice=my_frame[i]
    line="{"
    #process the specimen description
    for i in range(len(Specimen_description)):
        Key=Specimen_description[i]
        if attribute_types[Key]=="int" or attribute_types[Key]=="float":
            attribute=str(curr_slice[Specimen_description[i]])
        elif attribute_types[Key]=="dateTime":
            if str(curr_slice[Specimen_description[i]])=="NA" or str(curr_slice[Specimen_description[i]])=="None":
                attribute="\"None\""
            else:
                attribute="{$date:\""+str(curr_slice[Specimen_description[i]])+"\"}"
        else:
            attribute="\""+str(curr_slice[Specimen_description[i]])+"\""
        line=line+"\""+Key+"\":"+attribute+","

    #print("#### "+line)
    
    #process the Location description
    line=line+"\"Location\":{"
    for i in range(len(Location)):
        Key=Location[i]
        if attribute_types[Key]=="int" or attribute_types[Key]=="float":
            attribute=str(curr_slice[Location[i]])
        elif attribute_types[Key]=="dateTime":
            if str(curr_slice[Location[i]])=="NA" or str(curr_slice[Location[i]])=="None":
                attribute="\"None\""
            else:
                attribute="{$date:\""+str(curr_slice[Location[i]])+"\"}"
        else:
            attribute="\""+str(curr_slice[Location[i]])+"\""
        line=line+"\""+Key+"\":"+attribute
        
        if i==len(Location)-1:
            line=line+"},"
        else:
            line=line+","
    #print("#### "+line)

    #process the Collection description
    line=line+"\"Collection\":{"
    for i in range(len(Collection)):
        Key=Collection[i]
        if attribute_types[Key]=="int" or attribute_types[Key]=="float":
            attribute=str(curr_slice[Collection[i]])
        elif attribute_types[Key]=="dateTime":
            if str(curr_slice[Collection[i]])=="NA" or str(curr_slice[Collection[i]])=="None":
                attribute="\"None\""
            else:
                attribute="{$date:\""+str(curr_slice[Collection[i]])+"\"}"
        else:
            attribute="\""+str(curr_slice[Collection[i]])+"\""
        line=line+"\""+Key+"\":"+attribute
        
        if i==len(Collection)-1:
            line=line+"}"
        else:
            line=line+","
    
    
    line=line+"}"
    fout.write(line+"\n")
    print(line)
    
```
